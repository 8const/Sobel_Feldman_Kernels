```python
import numpy
from matplotlib.pyplot import imshow
import numpy as np
import matplotlib.pyplot as plt
from cv2 import imread, IMREAD_GRAYSCALE
```

```python
#Sobel–Feldman operator
kernel = np.array([[1, 0,-1], 
                   [2, 0,-2], 
                   [1, 0,-1]])

plt.subplot(1, 2, 1)
imshow(kernel, cmap='gray')
plt.subplot(1, 2, 2)
imshow(kernel.T, cmap='gray')
```

```python
#will give values for each pixel in result after convolution
#local is a 3x3 submatrix from the input image
pixel = lambda local, kernel: (local * kernel).sum()
```

```python
img = imread("1.jpg", IMREAD_GRAYSCALE)

#result.shape should be smaller by 2
result = np.zeros_like(img)[2:, 2:]
result.shape
```

```python
#gets 3x3 submatrix i steps down and j steps right counting from 0
#Moore neighborhood
get_local = lambda image, i, j: image[i-1:i+2, j-1:j+2]

#making a nice 2d array for testing with some texture
sq = np.concatenate([np.outer(np.arange(6), np.ones(3)), np.outer(np.ones(6), np.arange(3))], axis=1)
sq[3] = np.zeros(6)
plt.figure(figsize=(16,16))
ax = plt.subplot(1, 2, 1)
ax.set_title("sq")
imshow(sq, cmap="Pastel2")

ax = plt.subplot(1, 2, 2)
imshow(get_local(sq, 3,3), cmap="GnBu")
ax.set_title("get_local(sq, 3, 3)")
```

```python
#colors are different because they are chosen relatively to other pixels on pic
#so only texture is preserved
k = 1
plt.figure(figsize=(16,16))
for i in range(1, 5):
    for j in range(1,5):
        ax = plt.subplot(4,4,k)
        ax.set_title("get_local(sq,"+str(i)+", " + str(j)+")")
        imshow(get_local(sq, i,j), cmap='Pastel2')
        plt.axis('off')
        k+=1

#we see sliding 3x3 window sliding through the texture of sq
#result will be generated by applying pixel function to each 3x3 chunk
#because of border conditions the result is going to be 4x4 and not 6x6 as sq was
```

```python
#sanity check
#expecting value close to 0

ax = plt.subplot(1, 2, 1)
ax.set_title("get_local(sq, 1, 1)")

imshow(get_local(sq, 1, 1), cmap="gray")

ax = plt.subplot(1, 2, 2)
imshow(kernel, cmap="gray")
ax.set_title("Sobel–Feldman Kernel")

pixel(get_local(sq, 1, 1), kernel)
```

```python
#sanity check
#expecting negative value

ax = plt.subplot(1, 2, 1)
ax.set_title("get_local(sq, 1, 1)")

imshow(get_local(sq, 1, 1), cmap="gray")

ax = plt.subplot(1, 2, 2)
imshow(kernel.T, cmap="gray")
ax.set_title("Sobel–Feldman Kernel.T")

pixel(get_local(sq, 1, 1), kernel.T)
```

```python
#boring people know it's actually cross-correlation
def convolution(image, kernel, thr):
    result = np.zeros_like(img)[2:, 2:]
    for i in range(0, result.shape[0]):
        for j in range(0, result.shape[1]):
            p = pixel(get_local(img, i+1, j+1), kernel)
            if p > thr:
                result[i,j] = p
    return result
```

```python
Gx = convolution(img, kernel, 128)
Gy = convolution(img, kernel.T, 128)
G = np.sqrt(np.square(Gx/255) + np.square(Gy/255))
```

```python
plt.figure(figsize=(16,16))
plt.subplot(1,2,1)
imshow(img, cmap='gray')
plt.subplot(1,2,2)
imshow(G,cmap='gray')
```
